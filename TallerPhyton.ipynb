```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('Tabla-de-Análisis-de-datos-con-Python-..csv')

# Determinar cuántas encuestas se realizaron por zona de la ciudad
s_madrid = df.loc[df['ZONA'].str.contains('MADRID'), 'ENCUESTA'].sum()
s_facatativa = df.loc[df['ZONA'].str.contains('FACATATIVA'), 'ENCUESTA'].sum()
s_funza = df.loc[df['ZONA'].str.contains('FUNZA'), 'ENCUESTA'].sum()
s_mosquera = df.loc[df['ZONA'].str.contains('MOSQUERA'), 'ENCUESTA'].sum()

# Mostrar los resultados
print("Las encuestas realizadas en MADRID fueron:", s_madrid)
print("Las encuestas realizadas en FACATATIVA fueron:", s_facatativa)
print("Las encuestas realizadas en FUNZA fueron:", s_funza)
print("Las encuestas realizadas en MOSQUERA fueron:", s_mosquera)

# Crear una lista con los resultados específicos de cada zona y sus etiquetas
resultados_zona = [s_madrid, s_facatativa, s_funza, s_mosquera]
zonas = ['MADRID', 'FACATATIVA', 'FUNZA', 'MOSQUERA']

# Si quieres visualizarlo con un gráfico de barras, puedes hacerlo así:
plt.bar(zonas, resultados_zona, color='skyblue')
plt.xlabel('Zona de la Ciudad')
plt.ylabel('Número de Encuestas')
plt.title('Número de Encuestas por Zona de la Ciudad')
plt.ylim(0, 100)

# Mostrar los resultados individuales en el gráfico sin usar un bucle for
plt.text(zonas.index('MADRID'), s_madrid + 1, str(s_madrid), ha='center', va='bottom', color='black')
plt.text(zonas.index('FACATATIVA'), s_facatativa + 1, str(s_facatativa), ha='center', va='bottom', color='black')
plt.text(zonas.index('FUNZA'), s_funza + 1, str(s_funza), ha='center', va='bottom', color='black')
plt.text(zonas.index('MOSQUERA'), s_mosquera + 1, str(s_mosquera), ha='center', va='bottom', color='black')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('Tabla-de-Análisis-de-datos-con-Python-..csv')

# Determinar cuántas encuestas se realizaron por estado civil
s_soltero = df.loc[df['ESTADO CIVIL'].str.contains('S'), 'ENCUESTA'].sum()
s_casado = df.loc[df['ESTADO CIVIL'].str.contains('C'), 'ENCUESTA'].sum()
s_viudo = df.loc[df['ESTADO CIVIL'].str.contains('V'), 'ENCUESTA'].sum()

# Mostrar los resultados
print("Las encuestas realizadas a personas solteras fueron:", s_soltero)
print("Las encuestas realizadas a personas casadas fueron:", s_casado)
print("Las encuestas realizadas a personas viudas fueron:", s_viudo)

# Crear una lista con los resultados específicos de cada estado civil y sus etiquetas
resultados_estado_civil = [s_soltero, s_casado, s_viudo]
estados_civiles = ['SOLTERO', 'CASADO', 'VIUDO']

# Si quieres visualizarlo con un gráfico de barras, puedes hacerlo así:
plt.bar(estados_civiles, resultados_estado_civil, color='skyblue')
plt.xlabel('Estado Civil')
plt.ylabel('Número de Encuestas')
plt.title('Número de Encuestas por Estado Civil')
plt.ylim(0, 120)

# Mostrar los resultados individuales en el gráfico sin usar un bucle for
plt.text(estados_civiles.index('SOLTERO'), s_soltero + 1, str(s_soltero), ha='center', va='bottom', color='black')
plt.text(estados_civiles.index('CASADO'), s_casado + 1, str(s_casado), ha='center', va='bottom', color='black')
plt.text(estados_civiles.index('VIUDO'), s_viudo + 1, str(s_viudo), ha='center', va='bottom', color='black')

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('Tabla-de-Análisis-de-datos-con-Python-..csv')

# Determinar cuántas encuestas se realizaron por sexo y por zona de la ciudad
conteo_sexo_por_zona = df.groupby(['SEXO', 'ZONA'])['ENCUESTA'].sum().unstack()

# Mostrar los resultados
print("Número de encuestas por sexo y zona de la ciudad:")
print(conteo_sexo_por_zona)

# Crear un gráfico de barras
ax = conteo_sexo_por_zona.T.plot(kind='bar', rot=0, color=['skyblue', 'lightcoral'])
plt.xlabel('Zona de la Ciudad')
plt.ylabel('Número de Encuestas')
plt.title('Número de Encuestas por Sexo y Zona de la Ciudad')
plt.legend(title='Sexo', bbox_to_anchor=(1, 1))

# Mostrar los valores de cada barra sin utilizar un bucle for
for i, col in enumerate(conteo_sexo_por_zona.columns):
    for j, value in enumerate(conteo_sexo_por_zona[col]):
        ax.annotate(value, (i, value), ha='center', va='bottom', xytext=(0, 10), textcoords='offset points', color='black')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('Tabla-de-Análisis-de-datos-con-Python-..csv')

# Calcular la edad promedio por sexo y estado civil
edad_promedio_por_sexo_estado_civil = df.groupby(['SEXO', 'ESTADO_CIVIL'])['EDAD_PROMEDIO'].mean().unstack()


# Mostrar los resultados
print("Edad promedio por sexo y estado civil:")
print(edad_promedio_por_sexo_estado_civil)

# Crear un gráfico de barras para la edad promedio
ax_edad = edad_promedio_por_sexo_estado_civil.T.plot(kind='bar', rot=0, color=['skyblue', 'lightcoral'])
plt.xlabel('Estado Civil')
plt.ylabel('Edad Promedio')
plt.title('Edad Promedio por Sexo y Estado Civil')
plt.legend(title='Sexo', bbox_to_anchor=(1, 1))

# Mostrar los valores de cada barra sin utilizar un bucle for
for i, col in enumerate(edad_promedio_por_sexo_estado_civil.columns):
    for j, value in enumerate(edad_promedio_por_sexo_estado_civil[col]):
        ax_edad.annotate(f'{value:.2f}', (i, value), ha='center', va='bottom', xytext=(0, 10), textcoords='offset points', color='black')

plt.show()
```
